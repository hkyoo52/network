```python
import math
import matplotlib.pyplot as plt

x=[i/10 for i in range(0,34)]
y=[j*math.exp(-j) for j in x]
plt.plot(x,y)
plt.show()
```

```python
def get_prob(array,slot,p):
  prob=0
  for i in range(slot):
    count=0
    for j in array:
      if count==2:
        break
      if j[i]<p:
        count+=1
        continue
    if count==1:
      prob+=1
  return prob/slot
```

```python
import numpy as np
N=200
slot=1000
terminal=200
rand=np.random.rand(terminal,slot)
mu=[i/10 for i in range(0,34)]
prob=mu/np.full(np.array(mu).shape,terminal)
y=[]
for i in range(len(prob)):
  y.append(get_prob(rand,slot,prob[i]))
```

```python
plt.plot(mu,y)
plt.show()
```
